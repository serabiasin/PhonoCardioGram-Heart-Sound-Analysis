```
import pywt
import librosa
import numpy as np
```

```
sampelfile="/content/drive/My Drive/DataSetSuaraJantungPhysionet/challenge-2016/1.0.0/training-a/a0001.wav"

#read file,set to default SR from file
#note SR File Dataset default 2000 Hz
x,sr=librosa.load(sampelfile,sr=None)
x=x[:sr*10]
print("Sampling rate :",sr)
```

```
#test audio
import IPython.display as ipd
ipd.Audio(x,rate=sr) # load a NumPy array
```

```
#check frequency domain 
from scipy import signal
from scipy.fft import fftshift
import matplotlib.pyplot as plt

f, t, Sxx = signal.spectrogram(x, sr)
plt.pcolormesh(t, f, Sxx)
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```

```
#objective : 
# mencari rentang frekuensi 30 - 70 Hz
# localisasi waktu lebih diutamakan
# semakin kecil scale, semakin terlihat jelas perubahan terhadap waktu dibanding frekuensi

Motherwavelet=pywt.ContinuousWavelet('gaus4')
[coefficient,freq]=pywt.cwt(x,np.arange(1,64),wavelet=Motherwavelet)
print(coefficient.shape)
```

```
 #scale = 35 (JANGAN DI RUN!)
 #Sumbu X merupakan waktu
 #Sumbu Y merupakan frekuensi dalam koefisien wavelets
 plt.matshow(coefficient, aspect = 'auto')
 plt.show() 
```
